```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer  # text documents into a numerical representation
from sklearn.linear_model import LogisticRegression  # binary classification problems (two classes spam mail, ham mail)
from sklearn.metrics import accuracy_score
```

```python
spam_email = pd.read_csv("spam.csv")
```

```python
spam_email
```

```python
# repalce null values with null strings
clean_data = spam_email.where((pd.notnull(spam_email)), " ")
```

```python
clean_data.head()
```

```python
clean_data.shape
```

```python
#  label encoding

clean_data.loc[clean_data["Category"] == "spam", "Category"] = 0
clean_data.loc[clean_data["Category"] == "ham", "Category"] = 1
```

```python
clean_data["Category"]
```

```python
x = clean_data["Message"]
y = clean_data["Category"]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size= 0.2, random_state=3)
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
print(x.shape)
print(x_test.shape)
print(x_train.shape)
```

```python
# text data into numerical data 

feature_extraction = TfidfVectorizer(min_df = 1, stop_words = "english", lowercase = True) 

x_train_feature = feature_extraction.fit_transform(x_train)
x_test_feature  = feature_extraction.transform(x_test) 

# y train and y test in into integers

y_train = y_train.astype("int")
y_test = y_test.astype("int")
```

```python
print(x_train_feature)
```

```python
# Train model through logistic regression model

model = LogisticRegression()
```

```python
# train the logictic model through training data

model.fit(x_train_feature, y_train)
```

```python
# evaluation prediction

prediction_training_data = model.predict(x_train_feature)

accuracy_training_data = accuracy_score(y_train, prediction_training_data)
```

```python
print("Accuracy of training data : ", accuracy_training_data)
```

```python
prediction_test_data = model.predict(x_test_feature)

accuracy_test_data = accuracy_score(y_test, prediction_test_data)
```

```python
print("Accuracy of test data : ", accuracy_test_data)
```
